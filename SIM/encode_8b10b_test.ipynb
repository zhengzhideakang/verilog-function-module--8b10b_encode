```python
from encdec8b10b import EncDec8B10B

def test_encdec8b10b():
    # 表头添加 "No." 列
    print(f"{'No.':<5} | {'Input (Dec)':<12} | {'Input (Hex)':<12} | {'8b/10b (Binary)':<15} | {'8b/10b (Hex)':<12} | {'Running Disp'}")
    print("-" * 100)

    rd = 0
    idx = 0  # 序号计数器

    # 定义12个控制码（与SV测试平台一致）
    kcode_list = [
        0x1C, 0x3C, 0x5C, 0x7C, 0x9C, 0xBC,
        0xDC, 0xFC, 0xF7, 0xFB, 0xFD, 0xFE
    ]

    # 1. 测试数据码 0~255 (正向)
    for data_in in range(256):
        rd, encoded_10b = EncDec8B10B.enc_8b10b(data_in, rd)
        print(f"{idx:<5} | {data_in:<12} | 0x{data_in:02X}         | 0b{encoded_10b:010b}    | 0x{encoded_10b:03X}       | {'RD+' if rd == 1 else 'RD-'}")
        idx += 1

    # 2. 测试控制码（12个码循环3遍）
    for repeat in range(3):
        for kcode in kcode_list:
            # 第三个参数为1表示输入为K码
            rd, encoded_10b = EncDec8B10B.enc_8b10b(kcode, rd, 1)
            print(f"{idx:<5} | {kcode:<12} | 0x{kcode:02X}         | 0b{encoded_10b:010b}    | 0x{encoded_10b:03X}       | {'RD+' if rd == 1 else 'RD-'}")
            idx += 1

    # 3. 测试数据码 255~0 (倒序)
    for data_in in range(255, -1, -1):
        rd, encoded_10b = EncDec8B10B.enc_8b10b(data_in, rd)
        print(f"{idx:<5} | {data_in:<12} | 0x{data_in:02X}         | 0b{encoded_10b:010b}    | 0x{encoded_10b:03X}       | {'RD+' if rd == 1 else 'RD-'}")
        idx += 1

if __name__ == "__main__":
    test_encdec8b10b()
```
